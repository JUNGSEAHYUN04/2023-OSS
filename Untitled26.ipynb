## 상관관계 분석하기

```python
years = [2017, 2018, 2019, 2020, 2021]

for year in years:
    # 해당 연도의 데이터 추출
    columns_to_compare = [
        f'{year}_대기1종', f'{year}_대기3종', f'{year}_대기4종', f'{year}_대기5종'
    ]
    air_quality_columns = [col for col in df_air_quality.columns if col.startswith(str(year))]

    # 해당 연도의 데이터프레임 생성
    df_year = df_combined[['구분별'] + columns_to_compare + air_quality_columns]



    # '구분별' 열 제거
    df_year = df_year.drop(columns=['구분별'])


    correlation_matrix = df_year.corr()


    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
    plt.title(f' {year}년도 배출량과 대기측정지표간의 상관관계')
    plt.show()
```

```python
# 배출량과 오염물질 농도의 시간적 변화를 시각화
def plot_time_series(df, variable, years):
    plt.figure(figsize=(14, 10))
    for year in years:
        plt.plot(df[f'{year}_{variable}'], label=f'{year}')
    plt.title(f'Time Series of {variable}')
    plt.xlabel('Index')
    plt.ylabel(variable)
    plt.legend()
    plt.show()

variables = ['SO2', 'CO', 'NO2', 'PM10', 'PM2_5', 'O3', '대기1종', '대기3종', '대기4종', '대기5종']

# 변수별 시계열 플롯
for var in variables:
    plot_time_series(df_combined, var, years)
```

```python
# 상관관계가 높은 변수 간의 산점도 및 회귀선 시각화
def plot_scatter_with_regression(df, x_var, y_var):
    sns.lmplot(x=x_var, y=y_var, data=df, aspect=1.5, ci=None)
    plt.title(f'산점도 및 회귀선 {x_var} and {y_var}')
    plt.show()

# 연도별 상관관계가 높은 변수 간 산점도 및 회귀선 플롯
for year in years:
    df_year = df_combined[[f'{year}_대기5종', f'{year}_NO2']]
    df_year.columns = [col.split('_')[-1] for col in df_year.columns]

    plot_scatter_with_regression(df_year, '대기5종', 'NO2')
```
